#### Get sensor locations

```python
import pandas as pd

query = ("https://data.melbourne.vic.gov.au/resource/xbm5-bb4n.json")
raw_locs = pd.read_json(query)
raw_locs
```

```python
from location import Location
locations = list()
FOV = 20
PAD = 2
for index, row in raw_locs.iterrows():
    loc = Location(row["latitude"], row["longitude"], row["sensorid"])
    loc.fov = FOV + 2*PAD 
    locations.append(loc)
print(len(locations))
```

```python
import urllib.request
import os

def get_street_view(loc, heading):
    save_dir = "../street-view"
    base = "https://maps.googleapis.com/maps/api/streetview?size=640x640"
    latlng = "location={},{}".format(loc.lat, loc.lng)
    view = "heading={}&fov={}".format(heading, loc.fov)
    key = "key={}".format(API_KEY)
    pitch= "pitch=-5"
    url = "{}&{}&{}&{}&{}".format(base, latlng, view, pitch, key)
    output = os.path.join(save_dir,
                          "melbourne/LOC_{}_h{}.jpg".format(loc.lid, heading))
    if os.path.exists(output):
        os.remove(output)
    urllib.request.urlretrieve(url, output)

headings = [ang*FOV for ang in range(0,int(360/FOV))]
print(headings)

def get_street_views(loc, headings=[0, 60, 120, 180, 240, 300]):
    for heading in headings:
        get_street_view(loc, heading)
```

```python
get_street_views(locations[0], headings=headings)
```

### Get Counts

```python
query = ("https://data.melbourne.vic.gov.au/resource/mxb8-wn4w.json?year=2015&$limit=400000")
mel_cnt = pd.read_json(query)
mel_cnt
```

```python
mel = mel_cnt.groupby(['sensor_id']).sum()[['qv_market_peel_st']]
mel
```

```python
from yolo_log_parser import process_logs
import pandas as pd

log_file = "../data/ped_count_melbourne.log"
df = pd.DataFrame(process_logs(log_file))
gsv = pd.DataFrame({'ped_exp_gsv':df.groupby(by=['lid'])['person'].sum(), 
                    'bike':df.groupby(by=['lid'])['bicycle'].sum()})
gsv
```

```python
ped = mel.join(gsv)
ped
```

```python
import seaborn as sns; sns.set(color_codes=True)
ax = sns.jointplot(x="ped_exp_gsv", y="qv_market_peel_st", data=ped, kind="reg")
sns.plt.show()
ped.ped_exp_gsv.corr(ped["qv_market_peel_st"])
```
