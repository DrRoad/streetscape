# What we are going to do

We are going to get one (or more) points on each street in a city. From the points gathered, we will collect GSV images and then count the number of pedestrians. By doing that, we can get a whole picture of a city's walking behavior.

### Steps

1. Get city streets (osmnx)
2. Create observation points from each street
3. Collect GSVs for each point (GSV api)
4. Count the number of pedestrians from each observation point
5. Assign the value on street
6. Draw street map colored with the number of pedestrians


## Get observation points

```python
import osmnx as ox
%matplotlib inline
ox.config(log_console=True, use_cache=True)
```

```python
city = ox.graph_from_place('Haywards Heath, UK')
```

```python
fig, ax = ox.plot_graph(city)
```

```python
import geopandas as gpd
streets = [{'osm_u':edge[0], 'osm_v': edge[1], **edge[2]} for edge in city.edges(data=True)]
```

```python
# when an edge is a straightline between nodes, it doesn't have geometry. So we have to make one.
from shapely.geometry import Point, LineString
for st in streets:
    if 'geometry' not in st:
        u = Point(city.node[st['osm_u']]['x'], city.node[st['osm_u']]['y'])
        v = Point(city.node[st['osm_v']]['x'], city.node[st['osm_v']]['y'])
        st['geometry'] = LineString([u, v])
df = gpd.GeoDataFrame(streets)[['osmid', 'name', 'length', 'geometry']]
```

```python
df = df.assign(obs_point = lambda x: x.geometry.interpolate(0.5, normalized=True))
df.head()
```

```python
pdf = gpd.GeoDataFrame(df['obs_point'], geometry='obs_point', crs={'init':'epsg:4326'})
pdf.plot(ax=df.plot(color='gray', figsize=(18,6)), color='red')
```

## Get Street Views

```python
API_KEY = ''  
FOV = 20
PAD = 2

import urllib.request
import os

def get_street_views(lid, lat, lng):
    headings = [ang*FOV for ang in range(0,int(360/FOV))]
    for heading in headings:
        save_dir = "../street-view"
        base = "https://maps.googleapis.com/maps/api/streetview?size=640x640"
        latlng = "location={},{}".format(lat, lng)
        view = "heading={}&fov={}".format(heading, FOV + 2*PAD)
        key = "key={}".format(API_KEY)
        pitch= "pitch=-5"
        url = "{}&{}&{}&{}&{}".format(base, latlng, view, pitch, key)
        output = os.path.join(save_dir,
                              "haywords_heath/LOC_{}_h{}.jpg".format(lid, heading))
        if os.path.exists(output):
            os.remove(output)
        urllib.request.urlretrieve(url, output)
```

```python
%%time
count = 0
for index, row in df.iterrows():
    lid = index
    lat = row['obs_point'].y
    lng = row['obs_point'].x
    get_street_views(lid, lat, lng)
    print(count)
    count += 1
    if count > 10:
        break
```
