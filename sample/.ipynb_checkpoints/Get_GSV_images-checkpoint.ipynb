# Get NYC Dataset

get ped count dataset from NYC

```python
import pandas as pd

query = ("https://data.cityofnewyork.us/resource/cqsj-cfgu.json")
raw_data = pd.read_json(query)
raw_data
```

# Extract lat lang coordinates


```python
from location import Location
locations = list()
FOV = 20
PAD = 2
for index, row in raw_data.iterrows():
    geom = row["the_geom"]
    loc = Location(geom["coordinates"][1], geom["coordinates"][0], row["loc"])
    loc.loc = row["loc"]
    loc.fov = FOV + 2*PAD 
    locations.append(loc)
print(len(locations))
```
