## Get GSV Images

```python
import geopandas as gpd
import requests 
    
query = "http://services1.arcgis.com/cNVyNtjGVZybOQWZ/ArcGIS/rest/services/WalkingCounts/FeatureServer/0/query?where=1%3D1&outFields=*&returnGeometry=true&returnIdsOnly=false&returnCountOnly=false&f=pgeojson"
query += "&inSR=%7B%22wkid%22%3A+4283%7D"
query += "&outSR=%7B%22wkid%22%3A+4326%7D"
    
data = requests.get(query)
data
```

```python
fc = data.json()
features = fc["features"]
crs = fc["crs"]
gdf = gpd.GeoDataFrame.from_features(features, crs={'init':4326})
gdf
```

```python
from location import Location
locations = list()
FOV = 20
PAD = 2
for index, row in gdf.iterrows():
    geom = row["geometry"]
    loc = Location(geom.y, geom.x, row["Site_ID"])
    loc.loc = row["Site_ID"]
    loc.fov = FOV + 2*PAD 
    locations.append(loc)
print(len(locations))
```

### Get Counts

```python
import urllib.request
def get_df_from_xls(link):    
    socket = urllib.request.urlopen(link)
    xd = pd.ExcelFile(socket)
    df = xd.parse(xd.sheet_names[0], header=0)[['Total Count', 'Site ID']]
    return df
```

```python
links = ['http://www.cityofsydney.nsw.gov.au/__data/assets/excel_doc/0003/278580/WalkingCountSydney_Oct2016.xlsx',
         'http://www.cityofsydney.nsw.gov.au/__data/assets/excel_doc/0005/260933/WalkingCountSydney_Mar2016.xlsx',
         'http://www.cityofsydney.nsw.gov.au/__data/assets/excel_doc/0006/260934/WalkingCountSydney_Oct2015.xlsx',
         'http://www.cityofsydney.nsw.gov.au/__data/assets/excel_doc/0007/260935/WalkingCountSydney_Mar2015.xlsx'
        ]

cnt_dfs = list()
for link in links:
    cnt_dfs.append(get_df_from_xls(link))
```

```python
syd_cnt = pd.concat(cnt_dfs)
syd_cnt = syd_cnt.groupby(['Site ID']).sum()
syd_cnt
```

```python
from yolo_log_parser import process_logs
import pandas as pd

log_file = "../data/ped_count_sydney.log"
df = pd.DataFrame(process_logs(log_file))
gsv = pd.DataFrame({'ped_exp_gsv':df.groupby(by=['lid'])['person'].sum(), 
                    'bike':df.groupby(by=['lid'])['bicycle'].sum()})
gsv
```

```python
ped = syd_cnt.join(gsv)
ped = ped[ped["ped_exp_gsv"]>0]
```

```python
import seaborn as sns; sns.set(color_codes=True)
ax = sns.jointplot(x="ped_exp_gsv", y="Total Count", data=ped, kind="reg")
sns.plt.show()
ped.ped_exp_gsv.corr(ped["Total Count"])
```
